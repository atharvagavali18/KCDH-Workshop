**Alternatively one can uses concept of Axes for finer control on placement of different graphs**
A figure contains one or more axes - each axes corresponding to a (sub)plot
Each axes can contain multiple plots
**Positioning of plot elements uses a 0  1 coordinate system - 1 is the width/height of the figure.
Specify the position using a four-element list or tuple to set bottom-left and top-right bounds.**

```python
#Using axes. setting titles and labels and insert text

# h = height, w = width
h,w=8,4
fig = plt.figure(figsize=(h,w))
# --- Main Axes
ax = fig.add_axes([0.1, 0.1, #[bottom(h*0.1), left(w*0.1),
                  0.8, 0.8])# top(h*0.8), right(w*0.8)]
ax.set_title('Main Title')

ax.plot(x, np.sin(x), '-g', label='sin(x)')
ax.set_xlabel('Data points')
ax.set_ylabel('sin(x)');
```

```python
#Nested plots using axes and inserted text

h,w=8,4 # h = height, w = width of figure
fig = plt.figure(figsize=(h,w))
# --- Main Axes
ax = fig.add_axes([0.1, 0.1, #[distance of bottom left point from left axes(w*0.1), bottom axes(h*0.1),
                  0.8, 0.8]) # width of axes (w*0.8), height of axes(h*0.8), ]
ax.set_title('Main Axes with Insert Child Axes')

ax.plot(x, np.sin(x), '-g', label='sin(x)')
ax.set_xlabel('Data points')
ax.set_ylabel('sin(x)')
# if you only execute above code a single graph will be generated withou the insert

# --- Inserted Axes
ax = fig.add_axes([0.18, 0.18, 0.2, 0.2])
ax.plot(x, np.cos(x), '-b', label='cos(x)')
ax.set_xticks([]); #don't put xticks in insert
ax.set_ylabel('cos(x)');

plt.figtext(0.88, 0.15, #location of text in proportion to figure size
            '@copyleft',
            ha='right', va='bottom');  # horizontal/vertical alignment
```

**Scatterplots- takes 2D arrays of points with options for point size, color, and more.**

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
#use normally distributed random numbers, and scale to be within 0-10, then add an x-dependent value
y = np.random.randn(num_points) * 10 + x*gradient
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(x, y) # The equally sized x and y arrays provide the point coordinates

fig.suptitle('A Simple Scatterplot'); #title for entire figure, not just for an axes, though here there is only one axes
```

```python
# Adjusting point size and color
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x*gradient
fig, ax = plt.subplots(figsize=(8, 4))
colors = np.random.rand(num_points)
size = np.pi * (2 + np.random.rand(num_points) * 8) ** 2
ax.scatter(x, y, s=size, c=colors, alpha=0.5)  # alpha sets transparency
fig.suptitle('Scatterplot with Color and Size Specified');
```

```python
# Scatter plot with regression line
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x*gradient
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(x, y)
m, c = np.polyfit(x, y ,1) # fit 1-dimensional polynomial
ax.plot(x, m*x + c)
fig.suptitle('Scatterplot With Regression-line');
```

**seaborn library - provides high level commands on top of matplotlib**

```python
import seaborn as sns
sns.set()  #set seaborn style as default
plt.rcParams['figure.figsize'] = [4.8, 3.6]
```

```python
#pair plots: plot all pair of columns against each other
iris = sns.load_dataset("iris")
iris.head()
```

```python
for col in ['sepal_length', 'sepal_width']:
    plt.hist(iris[col], density=True, alpha=0.5)
```

```python
#scatterplot
sns.scatterplot(data=iris, x="petal_length", y="petal_width",
                hue="species")
```

```python
sns.pairplot(iris, hue='species', height=1.6);
```

```python
#Linear regression plot with seaborn
sns.lmplot(data=iris, x='petal_length', y='petal_width',
           height=4, aspect=2)
plt.tight_layout()
```

```python
#customize the regression plot
sns.lmplot(data=iris, x='petal_length', y='petal_width', height=4, aspect=2,
           scatter_kws={"color": "slategray"},
           line_kws={"linewidth": 4, "linestyle":'--',
                     "color": "seagreen"},
           markers='D',
           ci=99.5) ; #confidence interval
```

**Facet Grid - focus on some variables while having multiple plots in a grid for values of two other variables**

```python
titanic=pd.read_csv('/content/titanic.csv')
titanic.head()
```

```python
grid = sns.FacetGrid(titanic, row="Sex", col="Pclass", margin_titles=True)
grid.map(plt.hist, "Survived")
```

```python
#need to set bins
grid = sns.FacetGrid(titanic, row="Sex", col="Pclass", margin_titles=True)
grid.map(plt.hist, "Survived",bins=3)
```

```python
#focus on easy comparison across class - set density=True in grid.map
grid = sns.FacetGrid(titanic, row="Sex", col="Pclass", margin_titles=True)
grid.map(plt.hist, "Survived",bins=3,density=True)
```

```python
#joint regression on two variables is easy
with sns.axes_style(style='white'):
    sns.jointplot(x="Age", y="Survived",
            data=titanic, kind="reg")
```

```python
#mathplotlib mathmode
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0.0, 1.0, 100)
s = np.cos(4 * np.pi * t) + 2

fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True)
ax.plot(t, s)

ax.set_title(r'$\sum_{n=1}^\infty \frac{-e^{i\pi}}{2^n}$!', fontsize=16, color='r');
```
